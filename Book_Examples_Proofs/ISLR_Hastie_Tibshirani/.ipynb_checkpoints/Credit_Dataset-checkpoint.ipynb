# Handling Qualitative Predictors

Based on Section 3.3.1 of ISLR

## Balance vs. Gender

```python
# Create dummy variable for Gender and retain only Gender_Female
df_gen_fem = pd.get_dummies(df_orig, columns = ['Gender']).drop(columns = ['Gender_Male'])
display(df_gen_fem.head())

X = df_gen_fem[['Gender_Female']] # Predictor
y = df_gen_fem['Balance'] # Response

lr = LinearRegression()
lr.fit(X, y)
print(f"Intercept: {lr.intercept_:0.2f}")
print(f"Gender(Female) coefficient: {lr.coef_[0]:0.2f}")
```

<b>Interpretation:</b>

* Average credit card debt for males is \\$509.80
* Average credit card debt for females is \\$529.53 (\\$509.8 + \\$19.73)

```python
# Create dummy variable for Gender and retain only Gender_Male
df_gen_male = pd.get_dummies(df_orig, columns = ['Gender']).drop(columns = ['Gender_Female'])
display(df_gen_male.head())

X = df_gen_male[['Gender_Male']] # Predictor
y = df_gen_male['Balance'] # Response

lr = LinearRegression()
lr.fit(X, y)
print(f"Intercept: {lr.intercept_:0.2f}")
print(f"Gender(Male) coefficient: {lr.coef_[0]:0.2f}")
```

<b>Interpretation:</b>

* Without the interaction term, slope of balance vs. income is independent of whether the individual is 'Student' or 'Non_Student'
* With the interaction term, slope of balance vs. income is lower for students as compared to non-students. This suggests that increases in income are associated with smaller increases in credit card balance among students as compared to non-students.
