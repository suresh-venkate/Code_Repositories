# Estimate mean and standard deviation of a Gaussian distribution

# Author: Suresh Venkatesan

* **Reference:** Pattern Recognition and Machine Learning, Christopher Biship, Section 1.2.4, pp. 26 - 27, Equations (1.55) and (1.56).
* Maximum-Likelihood Estimation (MLE) is used to estimate mean and standard deviation of a Gaussian distribution based on a i.i.d dataset generated from the distribution. 

# Import required libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(color_codes = True)
%matplotlib inline
```

# Generate random samples

```python
mean = 10 # Mean of the Gaussian distribution
std_dev = 5 # Standard Deviation of the Gaussian distribution

rng = np.random.default_rng(seed = 12345) # Define numpy random generator
iid_data = rng.normal(mean, std_dev, 10000) # Generate 10000 i.i.d samples

# Plot histogram of iid_data
plt.figure(figsize = (8, 6))
sns.histplot(iid_data, kde = True)
plt.title("Histogram of iid_data", fontsize = 20)
plt.ylabel("Count", fontsize = 20)
plt.xlabel("Values", fontsize = 20)
plt.show()
```

# Perform MLE

```python
mean_mle = np.mean(iid_data) # MLE of mean based on eqn (1.55)
std_mle = np.std(iid_data) # MLE of standard deviation based on eqn (1.56)
print("MLE estimate of mean is %0.4f" %mean_mle)
print("MLE estimate of standard deviation is mean is %0.4f" %std_mle)
```
