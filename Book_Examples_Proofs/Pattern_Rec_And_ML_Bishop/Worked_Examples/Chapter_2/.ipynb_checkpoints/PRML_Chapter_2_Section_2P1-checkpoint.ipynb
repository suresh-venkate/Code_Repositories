# Plot beta distribution for different values of a and b

```python
mu = np.linspace(1e-3, 0.999, 50)

fig, ax = plt.subplots(2, 2, figsize = (14, 10))
fig.suptitle("Plot of beta distribution for diffent values of hyperparameters a and b", fontsize = 30)

a, b = 0.1, 0.1
beta_pdf = stats.beta.pdf(mu, a, b)
ax[0, 0].plot(mu, beta_pdf, label = 'a = 0.1, b = 0.1')
ax[0, 0].legend(fontsize = 20)

a, b = 1, 1
beta_pdf = stats.beta.pdf(mu, a, b)
ax[0, 1].plot(mu, beta_pdf, label = 'a = 1, b = 1')
ax[0, 1].legend(fontsize = 20)

a, b = 2, 3
beta_pdf = stats.beta.pdf(mu, a, b)
ax[1, 0].plot(mu, beta_pdf, label = 'a = 2, b = 3')
ax[1, 0].legend(fontsize = 20)

a, b = 8, 4
beta_pdf = stats.beta.pdf(mu, a, b)
ax[1, 1].plot(mu, beta_pdf, label = 'a = 8, b = 4')
ax[1, 1].legend(fontsize = 20)

plt.show()
```
