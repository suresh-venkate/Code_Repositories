# Import required libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats
%matplotlib inline
```

# Illustration of Central Limit Theorem

Ref: PRML, Fig 2.6, Pg. 79

```python
M = 100 # Number of independent uniform RV's to generate
num_samp = 10000 # Number of samples to draw from each distribution
rv_array = np.zeros((M, num_samp)) # Array to store samples drawn from all distributions
for ind in range(M):
    unif_samp = stats.uniform().rvs(num_samp) # Generate num_samp samples from a new Uniform distn.
    rv_array[ind] = unif_samp

N_list = [1, 2, 10, 25, 50, 100] # Number of independent distributions to average
num_rows = 3 # Number of rows to use for plotting
num_cols = 2 # Number of columns to use for plotting
fig, ax = plt.subplots(num_rows, num_cols, figsize = (14, 15))
fig.suptitle("Illustration of CLT", fontsize = 30)
for ind, N in enumerate(N_list):
    rv_array_N = rv_array[:N] # Select samples from N independent distributions
    rv_array_mean = np.mean(rv_array_N, axis = 0) # Average samples of N independent distns.
    ax_row = ind // num_cols
    ax_col = ind - ax_row * num_cols
    ax_curr = ax[ax_row][ax_col]
    sns.histplot(rv_array_mean, stat = 'density', ax = ax_curr, label = f'N = {N}');
    ax_curr.set_xlabel("Values", fontsize = 20)
    ax_curr.set_ylabel("PMF", fontsize = 20)
    ax_curr.legend(fontsize = 15)
plt.show()    
```
