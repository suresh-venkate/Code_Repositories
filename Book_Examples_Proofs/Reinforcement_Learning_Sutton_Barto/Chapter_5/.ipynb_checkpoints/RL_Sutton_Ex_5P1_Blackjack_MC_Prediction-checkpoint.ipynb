# Monte Carlo Prediction - Blackjack

* **Description:** Perform first-visit MC prediction for the Blackjack example
* **Reference:** Reinforcement Learning, An Introduction, Second Edition by Sutton, Barto
* **Section:** Section 5.1, Example 5.1, Pg. 93

# Import required libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import gym
from collections import defaultdict
from tqdm import tqdm
```

# Define Functions and Classes

## Function: bj_policies

```python
def bj_policies(state, pol_ind):
    if (pol_ind == 1):
        if (state[0] >= 20):
            action = 0 # Stick
        else:
            action = 1 # Hit
            
    return action
```

## Function: gen_episode

```python
def gen_episode(env, policy, pol_ind):
    
    '''
    Generate one episode of agent-environment interaction
    Arguments:
        env: Instance of environment to use for generating episode
        policy: Policy to use for generating episode
        pol_ind: Index of policy to use for generating episode
        
    Returns:
        states_list: List of states encountered in the episode (includes init_state)
        actions_list: List of actions take by agent at each time step
        rewards_list: List of rewards received by agent at each time step
    '''
    
    states_list = [] # Placeholder to store list of all states encountered in episode
    actions_list = [] # Placeholder to store list of all actions taken by agent
    rewards_list = [] # Placeholder to store list of rewards received at each step
        
    curr_state = env.reset() # Initialize environment to a random starting state - S0
    while(1): # Run agent till episode is done
        states_list.append(curr_state) # Update states_list with current state  
        action = policy(curr_state, pol_ind) # Action taken by agent at time t
        next_state, rew, done, _ = env.step(action) 
        actions_list.append(action) # Update actions_list
        rewards_list.append(rew) # Update rewards list  
        curr_state = next_state
        if (done):
            break
            
    return states_list, actions_list, rewards_list        
```

## Function: mc_prediction

```python
def mc_prediction(env, policy, pol_ind, num_ep, gamma, visit_type):
    
    '''
    Function to run MC prediction and estimate state value function
    Arguments:
        env: Instance of environment to use for generating episode
        policy: Policy for which SVF has to be estimated
        pol_ind: Index of policy to use for generating episode
        num_ep: Number of episodes to use for MC prediction algorithm
        gamma: Discounting factor for reward computation
        visit_type: 'first' for first-visit MC prediction,
                    'every' for every-visit MC prediction.
        
    Returns:
        svf_pi: State value function for policy pi
        state_count: Number of times each state is encountered during MC prediction
    '''
    
    # Initialize value function and count for all states to zero
    svf_pi = defaultdict(int) # State value function for given policy 'pi'
    state_count = defaultdict(int) # Number of times each state is encountered across episodes
    
    # Loop through 'num_ep' episodes
    for ep in tqdm(range(num_ep)):
        ep_states, ep_actions, ep_rewards = gen_episode(env, policy, pol_ind) # Generate one episode
        num_timesteps = len(ep_rewards) # Number of timesteps in current episode
        G = 0 # Initialize return to 0
        
        # Loop through each timestep of current episode
        for ind in range((num_timesteps - 1), -1, -1): 
            G = gamma * G + ep_rewards[ind] # Update return of current timestep
            curr_state = ep_states[ind] # Retrieve state of current timestep
            if (visit_type == 'first'): # For first-visit MC prediction
                if (curr_state not in ep_states[0:ind]):
                    state_count[curr_state] += 1
                    svf_pi[curr_state] += (G - svf_pi[curr_state]) / state_count[curr_state]
            elif (visit_type == 'every'): # For every-visit MC prediction
                state_count[curr_state] += 1
                svf_pi[curr_state] += (G - svf_pi[curr_state]) / state_count[curr_state]
                 
        
    return svf_pi, state_count
```

## Function: alpha_sch

```python
def alpha_sch(ep_num, alpha_dict):
    """
    Function to vary alpha (step-size parameter for TD algorithm) dynamically as a function of episode number
    
    Arguments:
        ep_num: Current episode number
        alpha_dict: Dictionaty with information required to compute current alpha value
        
    Returns:
        alpha_curr: Value of alpha to use for current episode
    """
    
    alpha_init = alpha_dict['alpha_init'] # Starting value of alpha (to use in episode 0)
    decay_param = alpha_dict['decay_param'] # Alpha decay parameter
    alpha_final = alpha_dict['alpha_final'] # Final value of alpha (Alpha will be clamped to this value on the lower side)
    
    log_ep_num = np.log10(ep_num + 1)
    alpha_curr = alpha_init * ((decay_param) ** log_ep_num)
    if (alpha_curr <= alpha_final):
        alpha_curr = alpha_final
    
    return alpha_curr
```

## Function: td0_prediction

```python
def td0_prediction(env, policy, pol_ind, num_ep, alpha_dict, gamma):
    
    '''
    Function to run TD(0) prediction and estimate state value function
    Arguments:
        env: Instance of environment to use for generating episode
        policy: Policy for which SVF has to be estimated
        pol_ind: Index of policy to use for generating episode
        num_ep: Number of episodes to use for MC prediction algorithm
        alpha_dict: Dictionaty with information required to compute current alpha value.                
        gamma: Discounting factor for reward computation
        
    Returns:
        svf_pi: State value function for policy pi
    '''
    
    # Initialize value function and count for all states to zero
    svf_pi = defaultdict(int) # State value function for given policy 'pi'
    
    # Loop through 'num_ep' episodes
    for ep in tqdm(range(num_ep)):
        curr_state = env.reset()
        alpha = alpha_sch(ep, alpha_dict)
        while(1): # Run agent till episode is done
            action = policy(curr_state, pol_ind) # Action taken by agent at time t
            next_state, rew, done, _ = env.step(action) 
            if (done):
                svf_pi[curr_state] = svf_pi[curr_state] + alpha * (rew - svf_pi[curr_state])
                break
            else:
                svf_pi[curr_state] = svf_pi[curr_state] + alpha * (rew + gamma * svf_pi[next_state] - svf_pi[curr_state])
            curr_state = next_state

    return svf_pi
```

# Run MC Prediction

```python
env = gym.make('Blackjack-v1', sab=True) # Define Blackjack environment
env.reset() # Reset Environment
num_ep = 500000 # Number of episodes to use for MC prediction
gamma = 1 # Discounting factor

# Run first-visit MC prediction algorithm
svf_pi, state_count = mc_prediction(env, bj_policies, 1, num_ep, gamma, 'first') 

# Extract SVF for usable and non-usable ace cases from svf_pi
svf_pi_arr_ace_not_usable = np.zeros((10, 10)) # Placeholder for SVF array with non-usable ace
svf_pi_arr_ace_usable = np.zeros((10, 10)) # Placeholder for SVF array with usable ace
# Update SVF arrays from svf_pi
for i in range(10):
    for j in range(10):
        key_f = ((i + 12), (j + 1), False)
        key_t = ((i + 12), (j + 1), True)
        svf_pi_arr_ace_not_usable[i, j] = svf_pi[key_f]
        svf_pi_arr_ace_usable[i, j] = svf_pi[key_t]

# Plot state-value function
fig, ax = plt.subplots(2, 1, figsize = (10, 20))
fig.suptitle("State-Value Function after 500,000 episodes", fontsize = 25)
ax[0] = sns.heatmap(svf_pi_arr_ace_not_usable, linewidth=0.5, annot = True, ax = ax[0]) # SVF for no-usable ace case
ax[0].set_title("No-usable Ace", fontsize = 20)
ax[0].invert_yaxis()
ax[0].set_xticklabels(range(1, 11, 1), fontsize = 20)
ax[0].set_yticklabels(range(12, 22, 1), fontsize = 20)
ax[0].set_xlabel("Dealer Showing Card", fontsize = 20)
ax[0].set_ylabel("Player Sum", fontsize = 20)

ax[1] = sns.heatmap(svf_pi_arr_ace_usable, linewidth=0.5, annot = True, ax = ax[1]) # SVF for usable ace case
ax[1].set_title("Usable Ace", fontsize = 20)
ax[1].invert_yaxis()
ax[1].set_xticklabels(range(1, 11, 1), fontsize = 20)
ax[1].set_yticklabels(range(12, 22, 1), fontsize = 20)
ax[1].set_xlabel("Dealer Showing Card", fontsize = 20)
ax[1].set_ylabel("Player Sum", fontsize = 20);
```

# Run TD(0) Prediction

```python
env = gym.make('Blackjack-v1', sab=True) # Define Blackjack environment
env.reset() # Reset Environment
num_ep = 100000 # Number of episodes to use for MC prediction
gamma = 1 # Discounting factor
alpha_dict = {'alpha_init': 0.05,
              'decay_param': 0.9,
              'alpha_final': 1e-4
             }

# Run TD(0) prediction algorithm
svf_pi = td0_prediction(env, bj_policies, 1, num_ep, alpha_dict, gamma)

# Extract SVF for usable and non-usable ace cases from svf_pi
svf_pi_arr_ace_not_usable = np.zeros((10, 10)) # Placeholder for SVF array with non-usable ace
svf_pi_arr_ace_usable = np.zeros((10, 10)) # Placeholder for SVF array with usable ace
# Update SVF arrays from svf_pi
for i in range(10):
    for j in range(10):
        key_f = ((i + 12), (j + 1), False)
        key_t = ((i + 12), (j + 1), True)
        svf_pi_arr_ace_not_usable[i, j] = svf_pi[key_f]
        svf_pi_arr_ace_usable[i, j] = svf_pi[key_t]

# Plot state-value function
fig, ax = plt.subplots(2, 1, figsize = (10, 20))
fig.suptitle("State-Value Function after 500,000 episodes", fontsize = 25)
ax[0] = sns.heatmap(svf_pi_arr_ace_not_usable, linewidth=0.5, annot = True, ax = ax[0]) # SVF for no-usable ace case
ax[0].set_title("No-usable Ace", fontsize = 20)
ax[0].invert_yaxis()
ax[0].set_xticklabels(range(1, 11, 1), fontsize = 20)
ax[0].set_yticklabels(range(12, 22, 1), fontsize = 20)
ax[0].set_xlabel("Dealer Showing Card", fontsize = 20)
ax[0].set_ylabel("Player Sum", fontsize = 20)

ax[1] = sns.heatmap(svf_pi_arr_ace_usable, linewidth=0.5, annot = True, ax = ax[1]) # SVF for usable ace case
ax[1].set_title("Usable Ace", fontsize = 20)
ax[1].invert_yaxis()
ax[1].set_xticklabels(range(1, 11, 1), fontsize = 20)
ax[1].set_yticklabels(range(12, 22, 1), fontsize = 20)
ax[1].set_xlabel("Dealer Showing Card", fontsize = 20)
ax[1].set_ylabel("Player Sum", fontsize = 20);
```
