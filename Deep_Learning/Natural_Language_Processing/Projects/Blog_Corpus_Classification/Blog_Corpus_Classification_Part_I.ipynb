# Define required functions

## Function: Get list of all characters in a corpus

```python
def get_char_list(corpus):
    
    '''
    Get a list of lexicographically sorted characters present in corpus
    
    Arguments:
        corpus - Input corpus from which list of characters has to be extracted.
    
    Returns: 
        List of all characters in corpus sorted lexicographically.
    '''
    
    all_char_pattern = re.compile(r'.') # Match all characters in corpus except newline
    all_char_list = [] # Placeholder to store all characters in corpus

    for doc in tqdm(corpus): # Loop through all documents in corpus and extract all characters      
        doc = doc.lower() # Convert document to lower-case
        matches = all_char_pattern.finditer(doc)
        for mat in matches:
            all_char_list.append(mat.group(0)) # Append characters from current document

    # Get unique characters, convert to list and sort.
    uniq_char_list = sorted(list(set(all_char_list)))
    
    print()
    print("Number of unique characters is %d" %len(uniq_char_list))
    print()
    # print("List of unique characters:")
    # print("==========================")
    # print("{}".format(uniq_char_list))

    return uniq_char_list
```

## Function: Get list of all words in a corpus

```python
def get_words_list(corpus):
    
    '''
    Get a list of all words in the corpus sorted lexicographically. 
    
    Arguments:
        corpus - Input corpus from which list of words has to be extracted.
    
    Returns: 
        List of all words in corpus sorted lexicographically.
    '''
    
    all_words_list = [] # Placeholder to store all words in corpus

    for doc in tqdm(corpus): # Loop through all documents in corpus and extract all words
        doc = doc.lower() # Convert document to lower-case
        wordlist = doc.split(" ")
        all_words_list.extend(wordlist)

    uniq_words_list = sorted(list(set(all_words_list)))

    print()
    print("Number of unique words is %d" %len(uniq_words_list))
    print()
    # print("First 20 unique words:")
    # print("======================")
    # print("{}".format(uniq_words_list[:20]))
    # print()
    # print("Last 20 unique words:")
    # print("=====================")
    # print("{}".format(uniq_words_list[-20:]))

    return uniq_words_list
```

## Function: Pre-process features

```python
def preprocess_feat(corpus, stopword_list, stemmer):
    
    '''
    Pre-process each document in corpus. Includes the following:
        1. Convert to lower-case
        2. Remove html tags
        3. Delete single-quote characters (don't will become dont)
        3. Replace miscellaneous characters with whitespace
        4. Split document into words, perform stemming and merge stemmed words back into document
    
    Arguments:
        corpus - Input corpus.
        stemmer - Stemmer object to use for performing stemming
    
    Returns: 
        Pre-processed corpus.    
    '''

    corpus_docs_processed = [] # Placeholder to store processed documents from corpus
    for doc in tqdm(corpus): # Loop through all documents in corpus
        # Convert document to all lower-case
        doc_proc = doc.lower()

        # Search for html tags in the current document (if any) and replace them with space
        html_tag_patt = re.compile('<.*?>')
        doc_proc = re.sub(html_tag_patt, ' ', doc_proc)

        # Search for quote character (if any) and delete it
        # This will make words like don't = dont and won't = wont
        pattern_1 = re.compile(r'[\']')
        doc_proc = re.sub(pattern_1, '', doc_proc)

        # Search for miscellaneous characters (if any) and replace them with whitespace
        # Standard symbols in QWErtY keyboard to be replaced with whitespace
        pattern_2 = re.compile(r'[`|~|!|@|#|$|%|^|&|*|(|)|\-|_|=|+|\[|\]|{|}|\|;|:|"|,|.|/|<|>|?]')
        doc_proc = re.sub(pattern_2, ' ', doc_proc)

        # Split document into list of words, remove stopwords and perform stemming on each word
        words = [stemmer.stem(word) for word in doc_proc.split() if word not in stopword_list]  
        # Concatenate words back into a document
        doc_proc = ''
        for word in words:
            doc_proc = doc_proc + word + ' '

        # Append current processed document to corpus_docs_processed
        corpus_docs_processed.append(doc_proc)

    return corpus_docs_processed
```

## Function: Get results from fitted classifier

```python
def get_results(pipe, X_train, X_test, y_train, y_test):
    
    '''
    Function to extract results from fitted classifier pipe
    
    Arguments:
        pipe - Fitted classifier pipe
        X_train: Training set corpus docs
        X_test: Test set corpus docs
        y_train: Training set corpus labels
        y_test: Test set corpus labels
        
    '''    

    vect_fit = pipe.named_steps['vect'] # Extracted fitted vectorizer from pipe
    feat_names = vect_fit.get_feature_names() # Extract vocab from tfidf vectorizer
    print("Number of extracted features: {}".format(len(feat_names)))
    print()

    # Convert X_train and X_test into tf-idf vectors
    X_train_tfidf = vect_fit.transform(X_train)
    X_test_tfidf = vect_fit.transform(X_test)

    # Extract individual classifiers for gender, age, industry and astrosign
    clf_gender = pipe.named_steps['clf'].estimators_[0]
    clf_age = pipe.named_steps['clf'].estimators_[1]
    clf_industry = pipe.named_steps['clf'].estimators_[2]
    clf_astrosign = pipe.named_steps['clf'].estimators_[3]

    # Evaluate overall training and test set scores
    score_overall_train = pipe.score(X_train, y_train)
    score_overall_test = pipe.score(X_test, y_test)    
    
    # Evaluate each individual classifier on X_train_tfidf and X_test_tfidf
    score_gender_train = clf_gender.score(X_train_tfidf, y_train[:, 0])
    score_gender_test = clf_gender.score(X_test_tfidf, y_test[:, 0])
    score_age_train = clf_age.score(X_train_tfidf, y_train[:, 1])
    score_age_test = clf_age.score(X_test_tfidf, y_test[:, 1])
    score_industry_train = clf_industry.score(X_train_tfidf, y_train[:, 2])
    score_industry_test = clf_industry.score(X_test_tfidf, y_test[:, 2])
    score_astrosign_train = clf_astrosign.score(X_train_tfidf, y_train[:, 3])
    score_astrosign_test = clf_astrosign.score(X_test_tfidf, y_test[:, 3])

    print("Training set: Overall score: %0.2f %%" %(100 * score_overall_train))    
    print("Training set: Gender score: %0.2f %%" %(100 * score_gender_train))
    print("Training set: Age score: %0.2f %%" %(100 * score_age_train))
    print("Training set: Industry score: %0.2f %%" %(100 * score_industry_train))
    print("Training set: Astrosign score: %0.2f %%" %(100 * score_astrosign_train))
    print()
    print("Test set: Overall score: %0.2f %%" %(100 * score_overall_test))        
    print("Test set: Gender score: %0.2f %%" %(100 * score_gender_test))
    print("Test set: Age score: %0.2f %%" %(100 * score_age_test))
    print("Test set: Industry score: %0.2f %%" %(100 * score_industry_test))
    print("Test set: Astrosign score: %0.2f %%" %(100 * score_astrosign_test))
```

## Function: Define a Batch Generator

```python
def batch_generator(X, y_gen, y_age, y_ind, y_sgn, batch_size = 200):

    '''
    Batch generator to convert input from sparse array to dense numpy array.
    
    '''  
    
    while (True):

        # Generate indices for current batch and create current batch X and y
        indices = np.random.randint(0, X.shape[0], size = batch_size).tolist()
        batch_X = X[indices].toarray() # Convert sparse array to dense numpy array
        batch_y_gen = y_gen[indices]
        batch_y_age = y_age[indices]
        batch_y_ind = y_ind[indices]
        batch_y_sgn = y_sgn[indices]

        # Return batch - use yield function to make it a python generator
        yield batch_X, [batch_y_gen, batch_y_age, batch_y_ind, batch_y_sgn]
```

```python
# Common section for all outputs
# Inputs -> Dense(128, ReLU) -> Dropout(0.5) -> Dense(128, ReLU) -> Dropout(0.5) -> X
inputs = Input(shape = (max_features,))
X = Dense(128, activation = 'relu')(inputs)
X = Dropout(0.5)(X)
X = Dense(128, activation = 'relu')(X)
X = Dropout(0.5)(X)

# Branch for gender output
# X -> Dense(128, ReLU) -> Dropout(0.5) ->  Dense(64, ReLU) -> Dropout(0.5) ->
# -> Dense(64, ReLU) -> Dropout(0.5) -> Dense(1, sigmoid) -> gen_output
X_gen = Dense(128, activation = 'relu')(X)
X_gen = Dropout(0.5)(X_gen)
X_gen = Dense(64, activation = 'relu')(X_gen)
X_gen = Dropout(0.5)(X_gen)
X_gen = Dense(64, activation = 'relu')(X_gen)
X_gen = Dropout(0.5)(X_gen)
gen_output = Dense(1, activation = 'sigmoid', name = 'gen_op')(X_gen) # Gender label output

# Branch for age output
# X -> Dense(128, ReLU) -> Dropout(0.5) ->  Dense(64, ReLU) -> Dropout(0.5) ->
# -> Dense(64, ReLU) -> Dropout(0.5) -> Dense(3, softmax) -> age_output
X_age = Dense(128, activation = 'relu')(X)
X_age = Dropout(0.5)(X_age)
X_age = Dense(64, activation = 'relu')(X_age)
X_age = Dropout(0.5)(X_age)
X_age = Dense(64, activation = 'relu')(X_age)
X_age = Dropout(0.5)(X_age)
age_output = Dense(3, activation = 'softmax', name = 'age_op')(X_age) # Age label output

# Branch for industry output
# X -> Dense(128, ReLU) -> Dropout(0.5) ->  Dense(64, ReLU) -> Dropout(0.5) ->
# -> Dense(64, ReLU) -> Dropout(0.5) -> Dense(3, softmax) -> ind_output
X_ind = Dense(128, activation = 'relu')(X)
X_ind = Dropout(0.5)(X_ind)
X_ind = Dense(64, activation = 'relu')(X_ind)
X_ind = Dropout(0.5)(X_ind)
X_ind = Dense(64, activation = 'relu')(X_ind)
X_ind = Dropout(0.5)(X_ind)
ind_output = Dense(3, activation = 'softmax', name = 'ind_op')(X_ind) # Industry label output

# Branch for astro-sign output
# X -> Dense(128, ReLU) -> Dropout(0.5) ->  Dense(64, ReLU) -> Dropout(0.5) ->
# -> Dense(64, ReLU) -> Dropout(0.5) -> Dense(12, softmax) -> sgn_output
X_sgn = Dense(128, activation = 'relu')(X)
X_sgn = Dropout(0.5)(X_sgn)
X_sgn = Dense(64, activation = 'relu')(X_sgn)
X_sgn = Dropout(0.5)(X_sgn)
X_sgn = Dense(64, activation = 'relu')(X_sgn)
X_sgn = Dropout(0.5)(X_sgn)
sgn_output = Dense(12, activation = 'softmax', name = 'sgn_op')(X_sgn) # Sign label output

# Define overall model with single input and four outputs
model = Model(inputs = inputs, outputs = [gen_output, age_output, ind_output, sgn_output])
model.summary()
```

## Compile model

```python
# Define Optimizer - Adam with default settings
opt = optimizers.Adam() 

# Define Loss - BCE for gender and CCE for age, industry and astrosign
bce = keras.losses.BinaryCrossentropy() 
cce = keras.losses.CategoricalCrossentropy() 
loss_dict = {'gen_op': bce, 'age_op': cce, 'ind_op': cce, 'sgn_op': cce}

# Define Metric - Binary accuracy for gender and categorical accuracy for age, industry and astrosign
gen_met = keras.metrics.BinaryAccuracy()
age_met = keras.metrics.CategoricalAccuracy()
ind_met = keras.metrics.CategoricalAccuracy()
sgn_met = keras.metrics.CategoricalAccuracy()
met_dict = {'gen_op': [gen_met], 'age_op': [age_met], 'ind_op': [ind_met],'sgn_op': [sgn_met]}

model.compile(optimizer = opt, loss = loss_dict, metrics = met_dict)
```
