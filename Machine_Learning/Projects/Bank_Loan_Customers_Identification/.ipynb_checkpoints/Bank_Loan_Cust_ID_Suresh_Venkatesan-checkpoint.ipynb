# 2 Read the data as a data frame

```python
# Since the file is in .csv format, read_csv() command is used to read the data.
ploan_df = pd.read_csv('Bank_Personal_Loan_Modelling.csv')
display(ploan_df.head())
```

<font size="3"><b>There are 14 columns in the data:</b></font>

    1) ID: Customer ID. 
    2) Age: Age of Customer.
    3) Experience: Number of years of professional experience.
    4) Income: Annual income of customer in 1000's of dollars.
    5) Zip Code: Home address zip code.
    6) Family: Family size of the customer
    7) CCAvg: Average spending on credit cards per month in 1000's of dollars.
    8) Education: Education Level.
        1: Undergrad
        2: Graduate
        3: Advance / Professional
    9) Mortgage: Value of house mortgage, if any in 1000's of dollars.
    10) Personal Loan (target variable): Indicates whether the customer accepted the personal loan offered in the last campaign.
    11) Securities Account: Indicates whether the customer has a securities account with the bank.
    12) CD Account: Indicates whether the customer has a certificate of deposit (CD) account with the bank.
    13) Online: Indicates whether the customer uses internet banking facilities.
    14) CreditCard: Indicates whether the customer uses a credit card issued bythe bank.

<b>The response (output) variable is 'Personal Loan' which indicates whether the customer accepted the personal loan offered in the previous campaign.</b>

```python
# Drop ID column from dataframe since it will not be used for further processing
ploan_df.drop(columns = 'ID', inplace = True)
display(ploan_df.head())
```

## 2.1 Data Types of each attribute

```python
# Print the data types of all columns
ploan_df.dtypes
```

<font size="3"><b>The data types of the following attributes needs to be changed from 'int64' to 'category' since they need to be interpreted as 'categorical variables:'</b></font>

    1. Zip Code
    2. Family
    3. Education
    4. Personal Loan
    5. Securities Account
    6. CD Account
    7. Online
    8. CreditCard
    
    
<b>Note:</b> 'Family' is treated as a categorical variable here. It isnt very clear if larger family sizes will have larger rates of personal loan acceptance. Hence, we will treat it as a categorical variable with no significance to the ordering of family size.

```python
# Convert the above mentioned columns to 'category' datatype.
# Create list of columns to convert to type category
cols_to_cat = ['ZIP Code', 'Family', 'Education', 'Personal Loan',\
            'Securities Account', 'CD Account', 'Online', 'CreditCard']
for ind in cols_to_cat:
    ploan_df[ind] = ploan_df[ind].astype('category')

ploan_df.dtypes
```

* There is one ZIP Code value that is showing up as a 4-digit code (9307). Since all ZIP Codes in the US are 5-digit codes, we will convert this ZIP code to 93007.

## 6.1 Zip Code Mapper

* The dataset has a feature called "ZIP Code" which refers to the home address ZIP Code. 
* This feature has to be treated as a categorical variable and there are 467 unique values of this feature.
* One option would be to one-hot encode all the 467 levels. However, this may not be very meaningful as each individual zip-code will not have much predictive value. Also, several zipcodes appear in the dataset only once which also implies that one-hot encoding of the zip-code levels may not be a good idea.


* <b>The approach taken here is to derive two new features from "ZIP Code":</b>
    * <b>First new feature:</b> We first find the frequency of occurence of each zip code in the dataset. This is the number of times a particular zip code appears in the dataset. We then group the zip codes into multiple bins depending on their frequency of occurence. For instance, we could group all zip codes that appear less than 25 times in the dataset into a single group 'G1'. All zip codes that appear between 25 and 100 times could be grouped together into another group 'G2' and so on. This new feature will then be treated as a categorical feature and will have much lesser number of levels which can then be one-hot encoded.
    * <b>Second new feature:</b> For the second feature, we evaluate the percentage of customers within that zip code who accepted the personal loan offered in the last campaign. For e.g: if there are 20 users in a particular zip code and 5 of them accepted the personal loan , then the second feature will evaluate to 5/20 = 0.25 for that particular zip code. This feature will then be treated as a continuous numeric feature and passed to the model as such.

```python
def zip_code_mapper(df_orig, zc_col_name, zc_freq_col_name, zc_bins,\
                   zc_bin_labels, out_var, zc_score_col_name):
    
    """
    Function to map ZIP Codes to two new features.
    
    Arguments:
        df_orig: Name of the original dataframe containing ZIP Code column.
        zc_col_name: Name of the column containing ZIP Codes.
        zc_freq_col_name: Name of the column to store the 'First new feature'
        zc_bins: Bin edges to group together zip codes based on frequency
        zc_bin_labels: Labels to use for each group created based on zc_bins
        out_var: Name of the column containing the output variable. 
        zc_score_col_name: Name of the column to store the 'Second new feature'

    Returns:
        df_new: New dataframe containing the two new features. Original zip code column 
        will be dropped from this dataframe
    
    """
    
    zc_freq_df = df_orig[zc_col_name].value_counts().reset_index(name = zc_freq_col_name).\
    rename(columns = {'index' : zc_col_name})
    bins = zc_bins
    bin_labels = zc_bin_labels
    zc_freq_df[zc_freq_col_name] = pd.cut(zc_freq_df[zc_freq_col_name], bins, labels = bin_labels)
    df_new = df_orig.copy()
    df_new = pd.merge(df_orig, zc_freq_df, on = zc_col_name, how = 'left')
    
    zc_out_var_ct = pd.crosstab(df_orig[zc_col_name], df_orig[out_var], normalize = 'index')
    zc_out_var_ct.rename_axis(None, axis = 1, inplace = True)
    zc_out_var_ct.drop(columns = 0, inplace = True)
    zc_out_var_ct.rename(columns = {1 : zc_score_col_name}, inplace = True)
    zc_out_var_ct.reset_index(inplace = True)
    df_new = pd.merge(df_new, zc_out_var_ct, on = zc_col_name, how = 'left')
    df_new.drop(columns = zc_col_name, inplace = True)

    return df_new
```

## 6.2 Normalize numeric columns

* This function is used to fit different scalers to different columns and return the scaled columns for both the training and test sets in new dataframes. The original unscaled columns will be dropped in the the new dataframes.

```python
def norm_num_feat(X_train_us, X_test_us, scal_dict):
    
    """
    Function to normalize multiple columns of X_train_us and X_test_us using the data provided in
    scal_dict.
    
    Arguments:
        X_train_us - Training set features with unscaled numeric columns
        X_test_us - Test set features with unscaled numeric columns
        scal_dict: Dictionary of columns to be scaled and scalers to be applied on each column
                   Key of scal_dict = S.No of column to be scaled.
                   Each entry of clfs is a list with the name of the column to be scaled, 
                   name to use for the scaled column, type of the scaler.
                   Type of the scaler:
                       mm_sc refers to MinMaxScaler.
                       st_sc refers to StandardScaler.
                       rb_sc refers to RobustScaler.

    Returns:
        X_train_sc - Training set features with scaled numeric columns.
        X_test_sc - Test set features with scaled numeric columns.
    
    """
    
    # Create copies of the unscaled training and test sets.
    X_train_sc = X_train_us.copy()
    X_test_sc = X_test_us.copy()
    
    for ind in range(len(scal_dict)):
        us_col_name = scal_dict[ind][0]
        sc_col_name = scal_dict[ind][1]
        
        if scal_dict[ind][2] == 'mm_sc':
            scaler = MinMaxScaler()
        elif scal_dict[ind][2] == 'st_sc':
            scaler = StandardScaler()
        elif scal_dict[ind][2] == 'rb_sc':
            scaler = RobustScaler()
        
        scaler.fit(X_train_us[[us_col_name]])
        X_train_sc[sc_col_name] = scaler.transform(X_train_us[[us_col_name]])
        X_train_sc.drop(columns = us_col_name, inplace = True)
        X_test_sc[sc_col_name] = scaler.transform(X_test_us[[us_col_name]])
        X_test_sc.drop(columns = us_col_name, inplace = True)
        
    return X_train_sc, X_test_sc
```

## 6.3 Define mixed naive bayes model

* Before training all the models, let's define a custom mixed naive bayes model. 
* The reason for doing this is as follows: Scikit-Learn provides naive bayes models which can either handle binomial / multinomial features or numeric features which are assumed to have a Gaussian distribution. 
* In this case, we have a mix of binary features (including original binary features and one-hot encoded categorical features) and numeric features. 
* To handle this case, we will split the features into binary features and numeric features. We will fit the binary features using a Bernoulli Naive Bayes model and the numeric features using a Gaussian Naive Bayes model. 
* The posterior probabilities will then be extracted from both models and combined together to generate the overall posterior probability. 

```python
# Function to fit Mixed Naive Bayes model
def MixedNB_fit(X_train, y_train, nb_bin_cols, nb_num_cols):
    
    """
    Function to fit a mixed naive Bayes model.
    The 'nb_bin_cols' of X_train (binary columns of X_train) will be fit using Bernoulli NB.
    The 'nb_num_cols' of X_train (numeric columns of X_train) will be fit using Gaussian NB.
    
    Arguments:
        X_train - Training set features
        y_train - Training set output
        nb_bin_cols - List of binary columns to use with Bernoulli Naive Bayes
        nb_num_cols - List of numeric columns to use with Gaussian Naive Bayes
    
    Returns:
        bnb - Model fit with Bernoulli NB
        gnb - Model fit with Gaussian NB
    
    """
    # Separate binary and numeric columns in training set
    X_train_bin = X_train[nb_bin_cols]
    X_train_num = X_train[nb_num_cols]

    # Fit Bernoulli Naive Bayes model to binary features
    bnb = BernoulliNB().fit(X_train_bin, y_train)
    
    # Fit Gaussian Naive Bayes model to numeric features
    gnb = GaussianNB().fit(X_train_num, y_train)
    
    return bnb, gnb
```

```python
# Function to return probabilities and class outputs using Mixed Naive Bayes model
def MixedNB_pred(X, bnb, gnb, nb_bin_cols, nb_num_cols, pr_thresh = 0.5):
    
    """
    Function to return probabilities and class outputs using Mixed Naive Bayes model.
    Should be used only after calling MixedNB_fit on training set.
    
    Arguments:
        X - Features array
        bnb - Model fit using Bernoulli NB (output of MixedNB_fit)
        gnb - Model fit using Gaussian NB (output of MixedNB_fit)
        nb_bin_cols - List of binary columns to use with Bernoulli Naive Bayes
        nb_num_cols - List of numeric columns to use with Gaussian Naive Bayes
        pr_thresh - Probability threshold used to classify output as positive or negative class
    
    Returns:
        y_pred: Predicted class outputs corresponding to feature array X
        pr_c1: Overall posterior probability computed using the mixed naive bayes model.
        
    """
    # Separate binary and numeric columns in features array
    X_bin = X[nb_bin_cols]
    X_num = X[nb_num_cols]
    
    # Extract prior probability of y_train == 1
    pr_c1_prior = np.exp(bnb.class_log_prior_[1])

    # Compute posterior probabilities for the two models
    pr_c1_post_bnb = bnb.predict_proba(X_bin)[: ,1]
    pr_c1_post_gnb = gnb.predict_proba(X_num)[: ,1]    
   
    # Compute overall posterior probability for the mixed naive bayes model
    pr_c1 = pr_c1_prior * (pr_c1_post_bnb / pr_c1_prior) * (pr_c1_post_gnb / pr_c1_prior)
    
    y_pred = (pr_c1 >= pr_thresh).astype(int)
    
    return y_pred, pr_c1
```

## 6.7 Fit and evaluate all models

* This function will be used to fit all models (defined in 'clfs') to training set.
* All parameters listed in 'param_names' will then be evaluated on both training and test sets.
* Finally all the parameter outputs will be colllected into a DataFrame and printed to screen.

```python
def model_fit_eval(clfs, param_names, X_train, y_train, X_test, y_test, nb_bin_cols = None,\
                   nb_num_cols = None, print_train_res = False, print_test_res = False):

    """
    Function to fit all models to training set, evaluate all parameters on training and test sets
    All outputs are collected into a DataFrame and printed to screen
    
    Arguments:
        clfs - Dictionary of classifiers.
            Key of clfs = S.No of classifier
            Each entry of clfs is a list with the classifier object, name of the classifier 
            and type of the classifier.
        param_names - # List of parameter names that will be evaluated for each classifier.
        X_train - Training set features
        y_train - Training set output
        X_test - Test set features
        y_test - Test set output
        nb_bin_cols - List of binary columns to use with Bernoulli Naive Bayes
        nb_num_cols - List of numeric columns to use with Gaussian Naive Bayes
        print_train_res - If true, print summary of results on training set
        print_test_res - If true, print summary of results on test set        
    
    Returns:
        bnb - Model fit with Bernoulli NB
        gnb - Model fit with Gaussian NB
        model_summary_train_set - Dataframe with results summary of training set
        model_summary_test_set - Dataframe with results summary of test set
    
    """
    
    # Set flag to indicate if naive bayes classifier is executed
    nb_flag = 0
    
    # Initialize classifier names list. Will be used as index for output DataFrame
    clf_names = []

    # Array to store output parameters for each classifier for testing and training sets
    clf_params_train = np.zeros((len(clfs), len(param_names)))
    clf_params_test = np.zeros((len(clfs), len(param_names)))

    # Iterate through all classifiers in 'clfs'
    for ind in range(len(clfs)):
#        print(ind, end = ' ')
        clf_names.append(clfs[ind][1]) # Append classifier name to clf_names array
        
        # If classifier is 'Naive Bayes', fit using custom Mixed NB model
        if (clfs[ind][2] == 'NB'):
            nb_flag = 1
            bnb, gnb = MixedNB_fit(X_train, y_train, nb_bin_cols, nb_num_cols)
            y_train_pred, _ = MixedNB_pred(X_train, bnb, gnb, nb_bin_cols, nb_num_cols, 0.5)
            y_test_pred, _ = MixedNB_pred(X_test, bnb, gnb, nb_bin_cols, nb_num_cols, 0.5)
        else:
            clfs[ind][0].fit(X_train, y_train) # Fit classifier to training set
            y_train_pred = clfs[ind][0].predict(X_train) # Obtain predictions for training set
            y_test_pred = clfs[ind][0].predict(X_test) # Obtain predictions for test set
    
        # Compute output probabilities for training and test set samples for non-SVM classifiers
        # Compute output decision function for training and test set samples for SVM classifiers
        if (clfs[ind][2] == 'NB'):
            _, pr_c1_train = MixedNB_pred(X_train, bnb, gnb, nb_bin_cols, nb_num_cols, 0.5)
            _, pr_c1_test = MixedNB_pred(X_test, bnb, gnb, nb_bin_cols, nb_num_cols, 0.5)       
        elif (clfs[ind][2] == 'SVM'):
            df_train = clfs[ind][0].decision_function(X_train)
            df_test = clfs[ind][0].decision_function(X_test)
        else:
            pr_c1_train = clfs[ind][0].predict_proba(X_train)[:, 1]
            pr_c1_test = clfs[ind][0].predict_proba(X_test)[:, 1]
            
        # Compute confusion matrix components for training and testing sets
        cm_train = confusion_matrix(y_train, y_train_pred, labels = [1, 0])
        cm_test = confusion_matrix(y_test, y_test_pred, labels = [1, 0])
        
        TP_train = cm_train[0, 0]
        FN_train = cm_train[0, 1]
        FP_train = cm_train[1, 0]
        TN_train = cm_train[1, 1]
        
        TP_test = cm_test[0, 0]
        FN_test = cm_test[0, 1]
        FP_test = cm_test[1, 0]
        TN_test = cm_test[1, 1]
        
        # Compute evaluation metrics on training and testing sets
        # Classification Accuracy
        clf_params_train[ind, 0] = 100 * accuracy_score(y_train, y_train_pred)
        clf_params_test[ind, 0] = 100 * accuracy_score(y_test, y_test_pred)
        
        # Precision
        clf_params_train[ind, 1] = TP_train / (TP_train + FP_train)
        clf_params_test[ind, 1] = TP_test / (TP_test + FP_test)
        
        # Recall
        clf_params_train[ind, 2] = TP_train / (TP_train + FN_train)
        clf_params_test[ind, 2] = TP_test / (TP_test + FN_test)    
        
        # F1-Score
        clf_params_train[ind, 3] = f1_score(y_train, y_train_pred)
        clf_params_test[ind, 3] = f1_score(y_test, y_test_pred)  
        
        # AUC (Area under ROC curve)
        if (clfs[ind][2] != 'SVM'):
            clf_params_train[ind, 4] = roc_auc_score(y_train, pr_c1_train)
            clf_params_test[ind, 4] = roc_auc_score(y_test, pr_c1_test)        
        else:
            clf_params_train[ind, 4] = roc_auc_score(y_train, df_train)
            clf_params_test[ind, 4] = roc_auc_score(y_test, df_test)           
        
        # APS (Average Precision Score)
        if (clfs[ind][2] != 'SVM'):
            clf_params_train[ind, 5] = average_precision_score(y_train, pr_c1_train)
            clf_params_test[ind, 5] = average_precision_score(y_test, pr_c1_test)        
        else:
            clf_params_train[ind, 5] = average_precision_score(y_train, df_train)
            clf_params_test[ind, 5] = average_precision_score(y_test, df_test)  
        
    # Club results of all classifiers into a single dataframe
    model_summary_train_set = pd.DataFrame(clf_params_train,\
                                           index = clf_names, columns = param_names)
    model_summary_test_set = pd.DataFrame(clf_params_test,\
                                          index = clf_names, columns = param_names)
        
    if print_train_res:
        # Print summary of all models on training set
        print("Summary of all models on training set")
        print("=====================================")
        display(model_summary_train_set)
        print()
        
    if print_test_res:
        # Print summary of all models on test set        
        print("Summary of all models on test set")
        print("=====================================")
        display(model_summary_test_set)     
    
    # Initialize bnb and gnb to None if mixed naive bayes is not run
    if (nb_flag == 0):
        bnb = None
        gnb = None
    
    return bnb, gnb, model_summary_train_set, model_summary_test_set
```

## 6.8 Plot confusion matrix

* This function will be used to plot the confusion matrices for all classifier defined in 'clfs' evaluated on a dataset 'X' with true ouput labels given in 'y'.

```python
def plot_conf_mat(clfs, X, y, bnb, gnb, nb_bin_cols = None, nb_num_cols = None, num_cols = 3):

    """
    Function to plot confusion matrices for all classifiers for a given sample set
    and associated labels. All classifiers are expected to have been already fit before 
    running this function.
    
    Arguments:
        clfs - Dictionary of classifiers.
                Key of clfs = S.No of classifier
                Each entry of clfs is a list with the classifier object, name of the classifier 
                and type of the classifier.
        X - Feature array
        y - Class labels corresponding to input X
        bnb - Model fit with Bernoulli NB
        gnb - Model fit with Gaussian NB
        nb_bin_cols - List of binary columns to use with Bernoulli Naive Bayes
        nb_num_cols - List of numeric columns to use with Gaussian Naive Bayes
        num_cols - Number of confusion matrices to print in each row
    
    Returns:
        None
    
    """
    num_clfs = len(clfs) # Total number of classifiers
    
    # Number of rows of confusion matrices that need to be printed
    num_rows = (num_clfs // num_cols) + 1 
    
    plt.figure(figsize = (12, (3 * num_rows))) # Initialize figure to plot confusion matrices
    
    # Iterate through all classifiers in 'clfs'
    for ind in range(len(clfs)):
        # If classifier is 'Naive Bayes', fit using custom Mixed NB model
        if (clfs[ind][2] == 'NB'):
            y_pred, _ = MixedNB_pred(X, bnb, gnb, nb_bin_cols, nb_num_cols, 0.5)
        else:
            y_pred = clfs[ind][0].predict(X) # Obtain predictions corresponding to X
        cm = confusion_matrix(y, y_pred, labels = [1, 0])
        cm_df = pd.DataFrame(cm, index = ['A1', 'A0'], columns = ['P1', 'P0'])
    
        ax = plt.subplot(num_rows, num_cols ,(ind + 1))
        sns.heatmap(cm_df, annot_kws = {"fontsize": 15}, linewidths = 1,\
                    linecolor = 'black', cmap = 'Blues', annot = True ,fmt = 'g', cbar = False)
        ax.set_title(clfs[ind][1], fontsize = 13)
        ax.tick_params(labelsize = 12)
    
    plt.tight_layout()
    plt.show() 
```

## 6.9 Plot ROC curves

* This function will be used to plot the ROC curves for all classifier defined in 'clfs' evaluated on a dataset 'X' with true ouput labels given in 'y'.

```python
def plot_roc_curves(clfs, X, y, bnb, gnb, nb_bin_cols = None, nb_num_cols = None):

    """
    Function to plot ROC curves for all classifiers for a given sample set
    and associated labels. All classifiers are expected to have been already fit before 
    running this function.
    
    Arguments:
        clfs - Dictionary of classifiers.
                Key of clfs = S.No of classifier
                Each entry of clfs is a list with the classifier object, name of the classifier 
                and type of the classifier.
        X - Feature array
        y - Class labels corresponding to input X
        bnb - Model fit with Bernoulli NB
        gnb - Model fit with Gaussian NB          
        nb_bin_cols - List of binary columns to use with Bernoulli Naive Bayes
        nb_num_cols - List of numeric columns to use with Gaussian Naive Bayes
    
    Returns:
        None
    
    """
    num_clfs = len(clfs) # Total number of classifiers
    
    # Initialize figure, axes and color map to plot ROC curves
    plt.figure(figsize = (10, 10))
    ax1 = plt.subplot(1,1,1)
    color = plt.cm.jet(np.linspace(0, 1, num_clfs))
    ax1.set_prop_cycle(color = color)
    
    # Iterate through all classifiers in 'clfs'
    for ind in range(len(clfs)):
        if (clfs[ind][2] == 'NB'):
            _, y_score = MixedNB_pred(X, bnb, gnb, nb_bin_cols, nb_num_cols, 0.5)
        elif (clfs[ind][2] == 'SVM'):
            y_score = clfs[ind][0].decision_function(X)
        else:
            y_score = clfs[ind][0].predict_proba(X)[:, 1]
        
        fpr, tpr, _ = roc_curve(y, y_score)
        ax1.plot(fpr, tpr, label = clfs[ind][1])
        
    plt.xlabel("FPR")
    plt.ylabel("TPR (recall)")
    plt.legend(loc=0, fontsize = 14)
    plt.tight_layout()
    plt.show() 
```

## 6.11 Confusion Matrix - Default and Best Thresholds

* This function will be used to evaluate and print confusion matrix components with default and best thresholds (based on ROC curve)

```python
def pr_conf_mat_def_best(clfs, X, y, bnb, gnb, nb_bin_cols = None, nb_num_cols = None):
    
    """
    Function to evaluate confusion matrix components for all classifiers for a given sample set
    and associated labels. All classifiers are expected to have been already fit before 
    running this function. The confusion matrix will first be evaluated with the default probability
    and decision function thresholds. Then the best threshold will be obtained using the ROC plot
    and predictions will be made using this best threshold. The CM will then be evaluated again 
    and all the CM parameters for default and best thresholds will be printed.
    
    Arguments:
        clfs - Dictionary of classifiers.
                Key of clfs = S.No of classifier
                Each entry of clfs is a list with the classifier object, name of the classifier 
                and type of the classifier.
        X - Feature array
        y - Class labels corresponding to input X
        bnb - Model fit with Bernoulli NB
        gnb - Model fit with Gaussian NB          
        nb_bin_cols - List of binary columns to use with Bernoulli Naive Bayes
        nb_num_cols - List of numeric columns to use with Gaussian Naive Bayes
      
    
    Returns:
        None
    
    """  
    # Initialize classifier names list. Will be used as index for output DataFrame
    clf_names = []

    # Initialize array to store all confusion matrix parameters
    cm_param_names = ['TP_Def', 'FN_Def', 'FP_Def', 'TN_Def', 'Best_Thresh', 'TP_Best',\
                      'FN_Best', 'FP_Best', 'TN_Best']
    cm_params = np.zeros((len(clfs), len(cm_param_names))) 

    # Iterate through all classifiers in 'clfs'
    for ind in range(len(clfs)):
        clf_names.append(clfs[ind][1]) # Append classifier name to clf_names array
        
        if (clfs[ind][2] == 'NB'):
            _, y_score = MixedNB_pred(X, bnb, gnb, nb_bin_cols, nb_num_cols, 0.5)
        elif (clfs[ind][2] == 'SVM'):
            y_score = clfs[ind][0].decision_function(X)
        else:
            y_score = clfs[ind][0].predict_proba(X)[:, 1]
        
        fpr, tpr, thresholds = roc_curve(y, y_score)
        
        # Obtain best threshold that minimizes Euclidean distance between (1, 0) point 
        # on the ROC curve and all (TPR, FPR) values.
        dist_min_ind = np.argmin(((1 - tpr) ** 2) + (fpr ** 2))
        best_thresh = thresholds[dist_min_ind]
        
        # Obtain predictions with default threshold
        # If classifier is 'Naive Bayes', fit using custom Mixed NB model
        if (clfs[ind][2] == 'NB'):
            y_pred, _ = MixedNB_pred(X, bnb, gnb, nb_bin_cols, nb_num_cols, 0.5)
        else:
            y_pred = clfs[ind][0].predict(X) # Obtain predictions corresponding to X
    
        # Obtain predictions with best threshold
        # If classifier is 'Naive Bayes', fit using custom Mixed NB model
        if (clfs[ind][2] == 'NB'):
            y_pred_best, _ = MixedNB_pred(X, bnb, gnb, nb_bin_cols, nb_num_cols, best_thresh)
        elif (clfs[ind][2] == 'SVM'):
            y_pred_best = (clfs[ind][0].decision_function(X) >= best_thresh).astype(int)
        else:
            y_pred_best = (clfs[ind][0].predict_proba(X)[:, 1] >= best_thresh).astype(int)
            
        # Obtain confusion matrices corresponding to default and best thresholds
        cm_def = confusion_matrix(y, y_pred, labels = [1, 0])
        cm_best = confusion_matrix(y, y_pred_best, labels = [1, 0])
        
        cm_params[ind, 0] = cm_def[0, 0]
        cm_params[ind, 1] = cm_def[0, 1]
        cm_params[ind, 2] = cm_def[1, 0]
        cm_params[ind, 3] = cm_def[1, 1]
        
        cm_params[ind, 4] = best_thresh
        
        cm_params[ind, 5] = cm_best[0, 0]
        cm_params[ind, 6] = cm_best[0, 1]
        cm_params[ind, 7] = cm_best[1, 0]
        cm_params[ind, 8] = cm_best[1, 1]
        
    # Compile all parameters into a dataframe
    cm_params_summary_test_set = pd.DataFrame(cm_params,\
                                              index = clf_names, columns = cm_param_names)
    # Print dataframe
    display(cm_params_summary_test_set)
```

## 6.12 Hyperparameter Search

* This is a small helper function to vary hyperparameters for different classifiers and print the best results across hyperparameters.
* <b>Note:</b> No cross-validation is used here. Best hyperparameter is selected based on AUC metric on the test set.

```python
def hyper_par_search(clf_type, var1, var2, param_names, X_train, y_train, X_test, y_test,\
                     print_train_res = False, print_test_res = False, plot_res = False):

    """
    Function to run a classifier across different hyperparameters as specified in var1 and var2. 
    Extract classification accuracy, F1-Score, AUC and APS as a function of the hyperparameters.
    
    For KNN, var1 = number of neighbors
    For Logistic Regression, var1 = regularization parameter (C)
    For SVM with RBF kernel, var1 = regularization parameter (C), var2 = gamma
    For SVM with polynomial kernel, var1 = degree of polynomial, var2 = gamma   
    
    Arguments:
        clf_type - Type of classifier
        var1 - First hyperparameter
        var2 - Second hyperparameter
        param_names - List of parameter names that will be evaluated for each classifier.
        X_train - Training set features
        y_train - Training set output
        X_test - Test set features
        y_test - Test set output    
        print_train_res - If true, print summary of results on training set
        print_test_res - If true, print summary of results on test set          
        plot_res - If true, plot a few metrics across the hyperparameter values
   
    Returns:
        ms_train_set - Dataframe with results summary of training set
        ms_test_set - Dataframe with results summary of test set
    
    """
    
    clfs = {} # Initialize classifer dictionary
    clfs_array = [] # Initialize classifier array
    
    # Create classifier dictionary for all values in var1, var2
    if (clf_type == 'KNN'):
        for ind, v1 in enumerate(var1):
            clfs_array.append([ind, [KNeighborsClassifier(n_neighbors = v1),\
                              'KNN_num_neighbor_' + str(v1), clf_type]])
    elif (clf_type == 'linear'):
        for ind, v1 in enumerate(var1):
            clfs_array.append([ind,[LogisticRegression(class_weight = 'balanced',\
                                                       solver = 'liblinear', C = v1),\
                                    'LogReg_With_Bal_Cl_Weights_C_' + str("{:.3f}".format(v1)),\
                                    clf_type]])
    elif (clf_type == 'SVM_RBF'):
        plot_res = False
        for ind1, v1 in enumerate(var1):
            for ind2, v2 in enumerate(var2):
                ind = ind1 * len(var2) + ind2                
                clfs_array.append([ind,\
                                   [SVC(C = v1, kernel='rbf', gamma = v2, class_weight = 'balanced'),\
                                    'SVM_RBF_C_' + str("{:.3f}".format(v1))\
                                    + '_gamma_' + str("{:.3f}".format(v2)), 'SVM']])  
    elif (clf_type == 'SVM_Poly'):
        plot_res = False
        for ind1, v1 in enumerate(var1):
            for ind2, v2 in enumerate(var2):
                ind = ind1 * len(var2) + ind2   
                clfs_array.append([ind,\
                                   [SVC(C = 1000, kernel='poly', degree = v1, gamma = v2,\
                                        class_weight = 'balanced'),\
                                    'SVM_Poly_deg_' + str("{:.3f}".format(v1))\
                                    + '_gamma_' + str("{:.3f}".format(v2)), 'SVM']])  
                
    
            
    clfs = dict(clfs_array)

    # Evaluate all models in classifier dictionary and capture train and test 
    # results in two dataframes
    _, _, ms_train, ms_test = model_fit_eval(clfs, param_names, X_train, y_train, X_test, y_test,\
                          None, None, print_train_res, print_test_res)

    # Compute best values and corresponding classifier for the following parameters 
    max_param_list = ['Class_Acc(%)', 'F1-Score', 'AUC', 'APS']
    max_param_df = pd.concat([ms_test.idxmax()[max_param_list], ms_test.max()[max_param_list]], axis = 1)
    max_param_df.rename(columns = {0: 'Best_Value_Classifier', 1: 'Best_Value'}, inplace = True)
    display(max_param_df)
    
    if plot_res:
        fig_50 = plt.figure(50, figsize = (12,12))
        fig_50.suptitle('Classifier metrics across hyperparameters', fontsize = 20)
        
        ax1 = plt.subplot(2, 2, 1)
        ax1.plot(var1, ms_train['Class_Acc(%)'], 'b*-', label = 'Training_Set')
        ax1.plot(var1, ms_test['Class_Acc(%)'], 'r^-', label = 'Test_Set')
        if (clf_type == 'KNN'):
            ax1.set_xlabel('Number of Neighbors', fontsize = 20)
        elif (clf_type == 'linear'):
            ax1.set_xlabel('C_Regularization_Parameter', fontsize = 20)
        ax1.set_ylabel('Classification_Accuracy(%)',  fontsize = 20)
        ax1.legend(loc = 'best')
        
        ax2 = plt.subplot(2, 2, 2)
        ax2.plot(var1, ms_train['F1-Score'], 'b*-', label = 'Training_Set')
        ax2.plot(var1, ms_test['F1-Score'], 'r^-', label = 'Test_Set')
        if (clf_type == 'KNN'):
            ax2.set_xlabel('Number of Neighbors', fontsize = 20)
        elif (clf_type == 'linear'):
            ax2.set_xlabel('C_Regularization_Parameter', fontsize = 20)
        ax2.set_ylabel('F1-Score', fontsize = 20)
        ax2.legend(loc = 'best')
        
        ax3 = plt.subplot(2, 2, 3)
        ax3.plot(var1, ms_train['AUC'], 'b*-', label = 'Training_Set')
        ax3.plot(var1, ms_test['AUC'], 'r^-', label = 'Test_Set')
        if (clf_type == 'KNN'):
            ax3.set_xlabel('Number of Neighbors', fontsize = 20)
        elif (clf_type == 'linear'):
            ax3.set_xlabel('C_Regularization_Parameter', fontsize = 20)
        ax3.set_ylabel('AUC', fontsize = 20)
        ax3.legend(loc = 'best')
        
        ax4 = plt.subplot(2, 2, 4)
        ax4.plot(var1, ms_train['APS'], 'b*-', label = 'Training_Set')
        ax4.plot(var1, ms_test['APS'], 'r^-', label = 'Test_Set')
        if (clf_type == 'KNN'):
            ax4.set_xlabel('Number of Neighbors', fontsize = 20)
        elif (clf_type == 'linear'):
            ax4.set_xlabel('C_Regularization_Parameter', fontsize = 20)
        ax4.set_ylabel('APS', fontsize = 20)
        ax4.legend(loc = 'best')
        
        plt.tight_layout()
        plt.show()
    
    return ms_train, ms_test
```

Confusion matrix (CM) on test set reveals that no positives have been identified as expected. True Positives (TP) = False Positives (FP) = 0. There are 144 False Negative (FN) predictions. AUC = 0.5 as expected.
* <b>Logistic_Regression_With_Default_Settings:</b> This classifier shows improved classification accuracy but has very poor recall of only 63%. CM indicates that only 90 out of 144 actual positives have been predicted as belonging to the positive class which explains the poor recall. These results, however use the default probability threshold of > 0.5 to classify a sample as belonging to the positive class. By lowering this prediction threshold, we should be able to improve the number of TP's significantly and subsequently recall as well. However, this will worsen the FPR and Precision. Depending on the cost associated with falsely classifying a negative-class sample as belonging to the positive class, this may be okay from a business perspective. AUC score of 0.97 indicates that the recall can be significantly improved.
* <b>Logistic_Regression_With_Balanced_Class_Weigths:</b> Since the dataset is heavily skewed, using balanced class weights should improve the recall performance of the classifier. As expected, recall performance of this classifier is much better at 93%. CM indicates that 133 out of 144 actual positives have been correctly classified! (as compared to only 90 out of 144 with default logistic regression). There are only 11 FN's now. However, this comes as the cost of larger FP's (because the classifier is overall classifying more samples as positive). Subsequently, precision and classification accuracy are both lower than for 'logistic regression without class weights'. AUC score is now 0.97 which is only marginally better than logistic regression without class weights. If we adjust the probability thresholds carefully before making predictions, then we can expect that both versions of logistic regression (with and without class weights) will yield similar TPR's.
* <b>Mixed_Naive_Bayes:</b> This classifier shows very poor recall performance since it makes strong assumptions about the class-conditional independence of the input features. Only 81 out of 144 actual positives are correctly classified as positive. However, AUC score is still high at 0.96, and we should be able to get much better TPR by adjusting the prediction probability threshold.

* <b> KNN classifier:</b> This classifier is run for two values of K (number of nearest neighbors): K = 5 and K = 12. Both versions have very poor recall performance and K = 12 is worse than K = 5. Since the dataset output is heavily skewed towards the negative class, any test sample that we take is likely to see more neighboring training samples from the negative class as compared to the positive class because of which most of the samples will end up getting classified into the negative class. With K = 5, only 48 out of 144 actual positives are correctly classified as positive. Also, there are only 2 FP's indicating that the KNN classifier is heavily biased towards making negative class predictions. WIth K = 12, this effect is even more amplified. There are only 23 TP's now and zero FP's. K = 12 however, provides the best AUC performance (checked by doing a hyperparameter search for K in the range of 1 to 20 and using test set AUC as the metric to choose K). Overall, KNN has poorer AUC score than Logistic Regression / Mixed Naive Bayes.
* <b> Linear Support Vector Classifier:</b> The performance of this classifier with balanced class weights is very very close to the performance of the logistic regression classifier with balanced class weights. This is expected since both these classifiers can produce only linear decision boundaries. 
* <b> SVM with Poly and RBF Kernel:</b> Both the non-linear SVM classifiers perform similar to each other and the recall / AUC performance is only marginally better than the linear classifiers. However, both these classifiers are able to achieve higher precision, lower classification error rate, higher F1-score and higher APS than the linear classifiers indicating that these two non-linear classifiers are able to do well on classifying both the positive and negative classes. This in turn indicates that the actual decision boundary is non-linear and both the non-linear SVM classifiers are able to approximate the non-linear decision boundary better than the linear classifiers.

* <b> Random Forest and Gradient Boosted Classifiers:</b> Finally, both the ensemble methods show slightly poorer recall (with default thresholds) than the SVM models but have the best AUC, APS, classification accuracy, F1-scores of all the models. With properly adjusted prediction thresholds, it is expected that these two classifiers will perform the best of the lot.

## 15.3 SVM with RBF Kernel: C and gamma

```python
# Coarse search
# C is swept over the values [1, 10, 100, 1000]
# gamma is swept over the values [0.01, 0.1, 1, 10]
clf_type = 'SVM_RBF' # Type of classifier
var1 = list(np.power(10, np.linspace(0, 3, 4))) # Values for C - regularization parameter
var2 = list(np.power(10, np.linspace(-2, 1, 4))) # Values for gamma

ms_train, ms_test = hyper_par_search(clf_type, var1, var2, param_names, X_train, y_train,\
                     X_test, y_test, print_train_res = False, print_test_res = False,\
                                     plot_res = False)
```

<b>Based on coarse search results, we will set C = 1000 and search over gamma in the range (0.01, 1).</b>

```python
# Fine Search
# C is fixed at 1000.
# Gamma is swept over the values 0.01 to 1 in logarithmic steps with 10 points per decade.
clf_type = 'SVM_RBF' # Type of classifier
var1 = [1000] # Values for C - regularization parameter
var2 = list(np.power(10, np.linspace(-2, 0, 21))) # Values for gamma

ms_train, ms_test = hyper_par_search(clf_type, var1, var2, param_names, X_train, y_train,\
                     X_test, y_test, print_train_res = False, print_test_res = False,\
                                     plot_res = False)
```

<b>Best AUC is obtained for C = 1000 and gamma = 0.025.</b>
